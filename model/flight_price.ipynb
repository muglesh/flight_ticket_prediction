```python
import pandas as pd
```

```python
df1=pd.read_csv("Scraped_dataset.csv")
```

```python
df1
```

```python
def split(data):
    airline=[]
    Class=[]
    for i in data:
        temp=i.split("\n")
        airline.append(temp[0])
        Class.append(temp[-1])
    return pd.DataFrame({"airline":airline,"Class":Class})
temp1=split(df1["Airline-Class"])
df2=pd.concat([df1,temp1],axis=1)
df2.head()
```

```python
temp=[]
for i in range(len(df2)):
    temp.append(int(df2["Price"][i].replace(",","")))
df2["new_price"]=pd.Series(temp)
```

```python
df3=df2.drop(["Price","Airline-Class"],axis=1)
```

```python
df3.head()
```

```python
def split_loc(data):
    out=[]
    for i in data:
        temp=i.split("\n")
        out.append(temp[-1])
    return pd.Series(out)
destination=pd.DataFrame({"destination":split_loc(df3['Arrival Time'])})
departure=pd.DataFrame({"departure":split_loc(df3['Departure Time'])})

```

```python
df4=pd.concat([df3,destination,departure],axis=1)
```

```python
df4.head()
```

```python
def split_loc(data):
    out=[]
    for i in data:
        temp=i.split(":")[0]
        temp=int(temp)
        if(temp>=3 and temp<=12):
             out.append("morning")
        elif(temp>12 and temp<=18):
             out.append("afternoon")
        else:
             out.append("night")
    return out
boarding_time=pd.DataFrame(split_loc(df4["Departure Time"]))
df4["boarding_time"]=boarding_time
```

```python
df4.drop("Departure Time",axis=1,inplace=True)
```

```python
df4.head()
```

```python
from datetime import datetime,date 
def str_date(data):
    temp=[]
    for i in data:
        date_string=i.replace("/","")
        date_object = datetime.strptime(date_string, "%d%m%Y").date()
        temp.append(date_object)
    return temp
df4["booking"]=pd.Series(str_date(df4['Date of Booking']))
df4['journey']=pd.Series(str_date(df4['Date of Journey']))
```

```python
def numOfDays(date1, date2):
    if date2 > date1:
        return (date2 - date1).days
    else:
        return (date1 - date2).days
temp=[]
for i in range(len(df4)):
    value=numOfDays(df4.booking[i],df4.journey[i])
    temp.append(value)
    
```

```python
df4["days"]=pd.Series(temp)
df4.head()
```

```python
df5=df4.drop(["Date of Booking","Date of Journey","Arrival Time"],axis=1)
df5.head()
```

```python
df5.destination.unique()
```

```python
def hours(data):
    temp=[]
    for i in data:
        temp.append(int(i.split("h")[0]))
    return pd.Series(temp)
flying_hours=hours(df5.Duration)
df5["flying_hours"]=flying_hours
```

```python
clean=df5.drop(["Duration","booking","journey"],axis=1)
clean.head()
```

```python
clean["price"]=clean['new_price']
clean.drop("new_price",axis=1,inplace=True)
```

```python
clean['Total Stops'].unique()
```

```python
c=[i for i in clean['Total Stops'].unique() if i not in ['non-stop', '1-stop', '2+-stop'] ]
for i in c:
    clean['Total Stops'].replace(i,'1-stop',inplace=True)
clean['Total Stops'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder 
le= LabelEncoder()
clean["stops"]=abs(le.fit_transform(clean['Total Stops'])-2)
```

```python
clean.stops.replace(2,4,inplace=True)
clean.stops.replace(1,2,inplace=True)
clean.stops.replace(4,1,inplace=True)
clean[clean.stops==1]
```

```python
clean.to_csv('datanew.csv', encoding='utf-8')
```

```python
clean.drop("Total Stops",axis=1,inplace=True)
```

```python
clean_final=clean.copy()
clean_final
```

```python
clean_final.airline=le.fit_transform(clean_final.airline)
clean_final.Class=le.fit_transform(clean_final.Class)
clean_final.destination=le.fit_transform(clean_final.destination)
clean_final.departure=le.fit_transform(clean_final.departure)
clean_final.boarding_time=le.fit_transform(clean_final.boarding_time)
```

```python
price=clean_final.price
```

```python
clean_final.drop(["price",'flying_hours',"boarding_time"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(clean_final,price,test_size=0.2)
```

```python
   
from sklearn.tree import DecisionTreeRegressor 
 
# Create a decision tree regression model
dt = DecisionTreeRegressor()
 
# Fit the model to the data
dt.fit(x_train, y_train)


dt.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
 
# Create a random forest regression model
rf = RandomForestRegressor(n_estimators=20)

# Fit the model to the data
rf.fit(x_train, y_train)

rf.score(x_test,y_test)
```
